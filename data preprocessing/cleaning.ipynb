```python
import pandas as pd
import json

# Load the JSON data
with open('sample_data.json') as f:
    data = json.load(f)

# Flattening the data
flattened_data = []

for entry in data:
    company_name = entry["companyName"]
    risk_category = entry["riskCategory"]
    
    if entry["article"]:
        for article in entry["article"]:
            flattened_data.append({
                "publishedAt": article["publishedAt"],
                "author": article["author"],
                "urlToImage": article["urlToImage"],
                "description": article["description"],
                "source_name": article["source"]["name"],
                "source_id": article["source"]["id"],
                "title": article["title"],
                "url": article["url"],
                "content": article["content"],
                "companyName": company_name,
                "riskCategory": risk_category
            })
    else:
        flattened_data.append({
            "publishedAt": None,
            "author": None,
            "urlToImage": None,
            "description": None,
            "source_name": None,
            "source_id": None,
            "title": None,
            "url": None,
            "content": None,
            "companyName": company_name,
            "riskCategory": risk_category
        })

df_flattened = pd.DataFrame(flattened_data)

df_flattened.head()
```
