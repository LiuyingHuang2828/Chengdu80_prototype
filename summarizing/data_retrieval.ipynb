```python
import requests
```

```python
url = 'https://mqv9qer0fa.execute-api.ap-southeast-1.amazonaws.com/dev/fetch-news/tesla'

response = requests.get(url)

if response.status_code == 200:
    posts = response.json()
else:
    print('Error:', response.status_code)
    
posts = response.json()[0]['article']

posts
```

```python
print('Author of Article:', posts[0]['author'])
print('Description of Article:', posts[0]['description'])
print('Title of Article:', posts[0]['title'])
print('URL of Article:', posts[0]['url'])
print('URL of Image:', posts[0]['urlToImage'])
print('Published Date:', posts[0]['publishedAt'])
print('Content of Article:', posts[0]['content'])
```
